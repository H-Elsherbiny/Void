```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, SimpleImputer

threshold = 0.5
missing_ratio = Tess.isnull().mean()
cols_to_drop = missing_ratio[missing_ratio > threshold].index
Tess = Tess.drop(columns=cols_to_drop)
print(f"Dropped columns: {list(cols_to_drop)}")


num_cols = Tess.select_dtypes(include=["float64", "int64"]).columns
iter_imputer = IterativeImputer(random_state=42)
Tess[num_cols] = iter_imputer.fit_transform(Tess[num_cols])


cat_cols = Tess.select_dtypes(include=["object"]).columns
cat_imputer = SimpleImputer(strategy="most_frequent")
Tess[cat_cols] = cat_imputer.fit_transform(Tess[cat_cols])

Tess.isnull().sum()
```

```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder

X = Tess.drop(columns=["tfopwg_disp"])
y = Tess["tfopwg_disp"]
le = LabelEncoder()
y_encoded = le.fit_transform(y)
X = X.select_dtypes(include=["int64", "float64"])
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)


pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("xgb", XGBClassifier(
        objective="binary:logistic",
        eval_metric="logloss",
        use_label_encoder=False,
        random_state=42,

    ))
])

param_grid = {
    "xgb__n_estimators": [100, 200, 300],
    "xgb__max_depth": [3, 5, 7],
    "xgb__learning_rate": [0.01, 0.05, 0.1],
    "xgb__subsample": [0.7, 0.8, 1],
    "xgb__colsample_bytree": [0.7, 0.8, 1]
}


grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring="f1_weighted", n_jobs=-1, verbose=1)
grid_search.fit(X_train, y_train)


print("Best Parameters:", grid_search.best_params_)
print("Best F1-Weighted Score:", grid_search.best_score_)


y_pred = grid_search.predict(X_test)
print("Test Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
import pandas as pd
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import ExtraTreesClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report


extra_trees = ExtraTreesClassifier(random_state=42, class_weight="balanced")
gb = GradientBoostingClassifier(random_state=42)
lr = LogisticRegression(max_iter=500, class_weight="balanced", solver="liblinear")
knn = KNeighborsClassifier()
svc = SVC(probability=True, class_weight="balanced", random_state=42)

# Ensemble using VotingClassifier
ensemble = VotingClassifier(
    estimators=[
        ("et", extra_trees),
        ("gb", gb),
        ("lr", lr),
        ("knn", knn),
        ("svc", svc)
    ],
    voting="soft"
)

# Pipeline
pipeline = Pipeline([
    ("imputer", IterativeImputer(random_state=42)),
    ("scaler", StandardScaler()),
    ("ensemble", ensemble)
])

# Grid search parameters (example: you can expand)
param_grid = {
    "ensemble__et__n_estimators": [100, 200],
    "ensemble__gb__n_estimators": [100, 200],
    "ensemble__lr__C": [0.1, 1],
    "ensemble__knn__n_neighbors": [3, 5],
    "ensemble__svc__C": [0.1, 1]
}

# GridSearchCV
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=5,
    scoring="f1_weighted",
    n_jobs=-1,
    verbose=2
)

# Fit the model
grid_search.fit(X_train, y_train)

# Best parameters and score
print("Best Parameters:", grid_search.best_params_)
print("Best F1-Weighted Score:", grid_search.best_score_)

# Evaluate on test set
y_pred = grid_search.predict(X_test)
print("Test Classification Report:")
print(classification_report(y_test, y_pred))
```
